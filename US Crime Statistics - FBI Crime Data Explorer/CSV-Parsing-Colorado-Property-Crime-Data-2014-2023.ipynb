## Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Load Data

```python
file_path = 'All Property Crime Reported by Population_10-02-2024_Colorado.csv'
df = pd.read_csv(file_path)
```

## Step 3: Reshape Data (from wide to long format)

```python
df_long = pd.melt(df, id_vars=['series'], var_name='date', value_name='value')
```

## Step 4: Convert 'date' to datetime format

```python
df_long['date'] = pd.to_datetime(df_long['date'], format='%m-%Y')
```

## Step 5: Separate Data by Series

```python
colorado_data = df_long[df_long['series'] == 'Colorado']
clearance_data = df_long[df_long['series'] == 'Colorado Clearances']
```

## Step 6: Plot Data (Time Series)

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='value', data=colorado_data, label='Colorado', marker='o')
sns.lineplot(x='date', y='value', data=clearance_data, label='Colorado Clearances', marker='o')
plt.title('Property Crime Data over Time')
plt.ylabel('Number of Crimes')
plt.xlabel('Date')
plt.legend()
plt.show()
```
